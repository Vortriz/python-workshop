# Jupyter Pain Points Demo

This notebook demonstrates why traditional Jupyter notebooks can be frustrating for reproducible science and version control.

### 1. Hidden State (Out-of-order execution)

**Step 1:** Run the cell below to define `x = 10`.

```python
x = 10
```

**Step 2:** Run this cell to see the current value of `x`.

```python
print(f"The value of x is: {x}")
```

**Step 3:** Now, change `x = 10` to `x = 20` in the first code cell, but **DO NOT** run it. 

Then run the print cell again. It will still say 10! This is hidden state. In a complex notebook with 100 cells, these "ghost" variables make reproduction a nightmare.

### 2. Dependency Issues

Does the person you shared this with have `seaborn` or `scipy` installed? You won't know until they try to run this cell and it crashes. There is no way to specify dependencies inside this `.ipynb` file.

**The "Workaround":** Many people add `!pip install seaborn` inside the notebook. 

*Why this is bad:* It modifies your global (or current) environment behind the scenes, making it even harder to track what is installed where. It also doesn't solve the issue for someone who doesn't have internet access or is using a different package manager like Conda.

### 3. Git Hostility (Binary Blobs & JSON)

Open this file in a text editor. You'll see it's a massive JSON blob. The plot generated below is stored as a long base64-encoded string inside that JSON. 

If you change a single parameter in the plot, Git will show a diff that is thousands of lines long and completely unreadable.

```python
import numpy as np
import matplotlib.pyplot as plt
data = np.random.randn(10, 10)
plt.imshow(data)
plt.title("This image makes Git sad")
plt.show()
```
